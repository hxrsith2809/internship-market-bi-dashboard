```python
!pip install pandas numpy plotly

import pandas as pd
import numpy as np
import plotly.express as px

np.random.seed(7)

n = 1200

df = pd.DataFrame({
    "Job Title": np.random.choice(
        ["Data Analyst Intern", "Business Analyst Intern", "Product Analyst Intern", "Software Intern"], 
        n
    ),
    "Company": np.random.choice(
        ["Google", "Amazon", "Microsoft", "Indegene", "Deloitte", "PwC"], 
        n
    ),
    "Location": np.random.choice(
        ["New York", "California", "Texas", "Illinois", "Remote"], 
        n
    ),
    "Remote": np.random.choice(["Yes", "No"], n),
    "Salary": np.random.randint(22, 65, n) * 1000,
    "Month": np.random.choice(
        ["Jan", "Feb", "Mar", "Apr", "May", "Jun"], 
        n
    ),
    "Skill": np.random.choice(
        ["SQL", "Python", "Excel", "Power BI", "Tableau", "Statistics"], 
        n
    )
})

print("Total Jobs:", len(df))
print("Average Salary:", int(df["Salary"].mean()))
print("Remote Jobs (%):", round((df["Remote"] == "Yes").mean() * 100, 2))

px.bar(
    df.groupby("Job Title").size().reset_index(name="Count"),
    x="Job Title",
    y="Count",
    color="Job Title",
    title="Internship Jobs by Role",
    text="Count"
).show()

px.bar(
    df.groupby("Skill").size().reset_index(name="Count"),
    x="Skill",
    y="Count",
    color="Skill",
    title="Skill Demand Across Internships",
    text="Count"
).show()

px.pie(
    df,
    names="Remote",
    hole=0.5,
    title="Remote vs Onsite Internship Roles",
    color_discrete_sequence=px.colors.sequential.RdBu
).show()

px.box(
    df,
    x="Job Title",
    y="Salary",
    color="Job Title",
    title="Salary Distribution by Internship Role"
).show()

px.line(
    df.groupby("Month").size().reset_index(name="Count"),
    x="Month",
    y="Count",
    markers=True,
    title="Hiring Trends Over Time",
    color_discrete_sequence=px.colors.qualitative.Bold
).show()

px.bar(
    df.groupby("Company").size().reset_index(name="Count"),
    x="Company",
    y="Count",
    color="Company",
    title="Top Companies Hiring Interns"
).show()

px.treemap(
    df,
    path=["Job Title", "Skill"],
    title="Skill Distribution by Internship Role",
    color="Salary",
    color_continuous_scale="Viridis"
).show()

px.histogram(
    df,
    x="Salary",
    color="Job Title",
    nbins=30,
    title="Salary Distribution Across Roles"
).show()

px.sunburst(
    df,
    path=["Location", "Job Title"],
    title="Location and Role Breakdown"
).show()

print("Interactive BI dashboard generated successfully.")
```
